```python
# Use numpy to convert to arrays
import numpy as np

# Labels are the values we want to predict
labels = np.array(df1['TargetVariable'])

# Remove the labels from the features
# axis 1 refers to the columns
features= df1.drop('TargetVariable', axis = 1)

# Saving feature names for later use
feature_list = list(features.columns)

# Convert to numpy array
features = np.array(features)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)

# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
#Mean Absolute Error: 3.83 degrees.
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
evaluation = pd.DataFrame({'Model': [],
                           'Details':[],
                           'Root Mean Squared Error (RMSE)':[],
                           'R-squared (training)':[],
                           'Adjusted R-squared (training)':[],
                           'R-squared (test)':[],
                           'Adjusted R-squared (test)':[],
                           '5-Fold Cross Validation':[]})
```

```python
def adjustedR2(r2,n,k):
    return r2-(k-1)/(n-k)*(1-r2)
```

```python
train_data_dm,test_data_dm = train_test_split(df2,train_size = 0.8,random_state=42)
```

Multiple Regression

```python
features =  ['Pagelikes', 'PageCheckins', 'Pagetalkingabout', 'Feature6', 'Feature7',
       'Feature11', 'Feature12', 'Feature13', 'Feature14', 'Feature17',
       'Feature18', 'Feature21', 'Feature22', 'Feature23', 'Feature24',
       'Feature25', 'Feature27', 'Feature29', 'CC1', 'CC2', 'CC4', 'CC5',
       'BaseTime', 'PostLength', 'PostShareCount']
complex_model_1 = linear_model.LinearRegression()
complex_model_1.fit(train_data_dm[features],train_data_dm['TargetVariable'])

print('Intercept: {}'.format(complex_model_1.intercept_))
print('Coefficients: {}'.format(complex_model_1.coef_))

pred = complex_model_1.predict(test_data_dm[features])
rmsecm = float(format(np.sqrt(metrics.mean_squared_error(test_data_dm['TargetVariable'],pred)),'.3f'))
rtrcm = float(format(complex_model_1.score(train_data_dm[features],train_data_dm['TargetVariable']),'.3f'))
artrcm = float(format(adjustedR2(complex_model_1.score(train_data_dm[features],train_data_dm['TargetVariable']),train_data_dm.shape[0],len(features)),'.3f'))
rtecm = float(format(complex_model_1.score(test_data_dm[features],test_data_dm['TargetVariable']),'.3f'))
artecm = float(format(adjustedR2(complex_model_1.score(test_data_dm[features],test_data['TargetVariable']),test_data_dm.shape[0],len(features)),'.3f'))
cv = float(format(cross_val_score(complex_model_1,df2[features],df2['TargetVariable'],cv=5).mean(),'.3f'))

r = evaluation.shape[0]
evaluation.loc[r] = ['Multiple Regression-1','selected features',rmsecm,rtrcm,artrcm,rtecm,artecm,cv]
evaluation.sort_values(by = '5-Fold Cross Validation', ascending=False)
```

2. Mean of Residuals
Residuals as we know are the differences between the true value and the predicted value. One of the assumptions of linear regression is that the mean of the residuals should be zero. So let's find out

```python
residuals=train_data_dm['TargetVariable']-y_pred
p = sns.scatterplot(y_pred,residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

Residuals not following Homoscadecity

```python
from scipy.stats import bartlett
test = bartlett(train_data_dm['TargetVariable'],residuals)
print(test)
```

Since p value is quite less than 0.05 in Bartlett, it's null hypothesis that error terms are homoscedastic gets rejected, that's not good for a regression

```python
p = sns.distplot(residuals,kde=True)
p = plt.title('Normality of error terms/residuals')
```

Longer tail so not normally distributed.

```python
from scipy.stats import shapiro
shapiro(residuals)
```

P is zero, Shapiro test we reject null hypothesis , residuals are not normally distributed
